### 🐍 Python SDK

The simplest way of simply using YOLOv8 directly in a Python environment.

````python
import os
import glob
import random
from ultralytics import YOLO
from IPython.display import Image, display

# 1. Define HOME directory
HOME = os.getcwd()

# 2. Safety Check: Determine which model to load
# Your previous run failed in 'runs/obb/train2', but we should look for successful runs
model_path = f'{HOME}/runs/detect/train/weights/best.pt'

if os.path.exists(model_path):
    print(f"✅ Loading custom trained model: {model_path}")
    custom_model = YOLO(model_path)
else:
    print("⚠️ Custom weights not found! Falling back to pretrained YOLO11n.")
    # This allows you to test the script even if training hasn't finished
    custom_model = YOLO('yolo11n.pt')

# 3. Pick a random image
# Note: Ensure 'dataset.location' is defined from your Roboflow download cell
try:
    test_images_path = glob.glob(f"{dataset.location}/test/images/*.jpg")
    random_image_path = random.choice(test_images_path)
    print(f"Testing on: {random_image_path}")

    # 4. Run the predict command
    # task='detect' ensures we aren't using the broken OBB logic
    results = custom_model.predict(source=random_image_path, conf=0.25, save=True)

    # 5. Find the latest prediction folder
    # We look in 'runs/detect/predict' because that's the default for standard YOLO
    predict_folders = sorted(glob.glob(f'{HOME}/runs/detect/predict*'))

    if predict_folders:
        latest_predict_folder = predict_folders[-1]
        image_name = os.path.basename(random_image_path)
        result_image_path = os.path.join(latest_predict_folder, image_name)

        if os.path.exists(result_image_path):
            display(Image(filename=result_image_path, width=600))
        else:
            print(f"Could not find result image at {result_image_path}")
    else:
        print("No prediction folders found in runs/detect/")

except NameError:
    print("Error: 'dataset' variable not found. Make sure you ran the data download cell.")
except IndexError:
    print("Error: No images found in the test path. Check your dataset folder.")
````

````python
# Check if results are not empty
if len(results) > 0:
    result = results[0]

    # 1. Standard Box Coordinates (x_center, y_center, width, height)
    # Gagamit tayo ng .boxes instead of .obb
    print("Coordinates (xywh):")
    if result.boxes:
        print(result.boxes.xywh)
    else:
        print("No boxes detected.")

    print("\n-------------------\n")

    # 2. Confidence Score (Gaano ka-sure si AI?)
    print("Confidence Scores:")
    if result.boxes:
        print(result.boxes.conf)

    print("\n-------------------\n")

    # 3. Class IDs (0 = Hand, etc.)
    print("Class IDs:")
    if result.boxes:
        print(result.boxes.cls)
else:
    print("No results found.")
````
